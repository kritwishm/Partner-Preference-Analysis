```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
%matplotlib inline
```

```python
data = pd.read_csv('Partner Preference Survey - Response - Form responses 1.csv')
```

```python
data.head()
```

```python
data.columns
```

```python
sns.set_style('whitegrid')
```

```python
basic_params = ['Age', 'Height', 'Country Living in', 'Marital Status', 'Educational Qualification',
                'Occupation', 'Verified Annual Income', 'Religion', 'Mother Tongue']
```

```python
for param in basic_params:
    plt.figure(figsize=(10,5))
    graph = sns.countplot(data=data,x=param,hue='What is your gender?',palette='coolwarm_r')
    graph.set_title(param+' importance gender-wise', fontsize=15)
    for p in graph.patches:
            graph.annotate('{:.0f}'.format(p.get_height()), (p.get_x()+0.3, p.get_height()),
                           ha='center', va='bottom', color= 'black')
```

```python
lifestyle_params = ['Food Preference', 'Smoking Habit', 'Drinking Habit', 'Having Pets',
                    'Having Kids', 'Music Taste', 'Movie Taste', 'Family Background',
                    'Social Media Activity', 'Owning a personal car', 'Owning a House']
```

```python
for param in lifestyle_params:
    plt.figure(figsize=(10,5))
    graph = sns.countplot(data=data,x=param,hue='What is your gender?',palette='Paired_r')
    graph.set_title(param+' importance gender-wise',fontsize=15)
    for p in graph.patches:
            graph.annotate('{:.0f}'.format(p.get_height()), (p.get_x()+0.3, p.get_height()),
                           ha='center', va='bottom', color= 'black')
```

```python
basic_param_avg = pd.pivot_table(data,index=['What is your gender?'],values=basic_params,aggfunc='mean').T
```

```python
graph = basic_param_avg.plot(kind='bar',figsize=(16,6),title='Avg. Importance of Basic parameters')
for p in graph.patches:
    graph.annotate('{:.2f}'.format(p.get_height()), (p.get_x()+0.12, p.get_height()),
                   ha='center', va='bottom', color= 'black')
```

```python
lifestyle_param_avg = pd.pivot_table(data,index=['What is your gender?'],values=lifestyle_params,aggfunc='mean').T
```

```python
graph = lifestyle_param_avg.plot(kind='bar',figsize=(16,7),cmap='Accent',
                                 title='Avg. Importance of Lifestyle parameters')
for p in graph.patches:
    graph.annotate('{:.2f}'.format(p.get_height()), (p.get_x()+0.12, p.get_height()),
                   ha='center', va='bottom', color= 'black')
```
