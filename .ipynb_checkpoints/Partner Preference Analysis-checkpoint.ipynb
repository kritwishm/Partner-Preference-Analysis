```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
%matplotlib inline
```

```python
data = pd.read_csv('Partner Preference Survey - Response - Form responses 1.csv')
```

```python
data.head()
```

```python
gender_pivot = data.groupby('What is your gender?').agg('count')
```

```python
gender_pivot.rename(columns={'Age':'Gender'},inplace=True)
```

```python
gender_pivot.plot.pie(y='Gender',figsize=(8, 8))
```

```python
data.drop(data[data['What is your gender?']=='Prefer not to say'].index,inplace=True)
```

```python
sns.set_style('whitegrid')
```

```python
basic_params = ['Age', 'Height', 'Country Living in', 'Marital Status', 'Educational Qualification',
                'Occupation', 'Verified Annual Income', 'Religion', 'Mother Tongue']
```

```python
for param in basic_params:
    plt.figure(figsize=(10,5))
    graph = sns.countplot(data=data,x=param,hue='What is your gender?',palette='coolwarm_r')
    graph.set_title(param+' importance gender-wise', fontsize=15)
    for p in graph.patches:
            graph.annotate('{:.0f}'.format(p.get_height()), (p.get_x()+0.3, p.get_height()),
                           ha='center', va='bottom', color= 'black')
```

```python
lifestyle_params = ['Food Preference', 'Smoking Habit', 'Drinking Habit', 'Having Pets',
                    'Having Kids', 'Music Taste', 'Movie Taste', 'Family Background',
                    'Social Media Activity', 'Owning a personal car', 'Owning a House']
```

```python
for param in lifestyle_params:
    plt.figure(figsize=(10,5))
    graph = sns.countplot(data=data,x=param,hue='What is your gender?',palette='Paired_r')
    graph.set_title(param+' importance gender-wise',fontsize=15)
    for p in graph.patches:
            graph.annotate('{:.0f}'.format(p.get_height()), (p.get_x()+0.3, p.get_height()),
                           ha='center', va='bottom', color= 'black')
```

```python
basic_param_avg = pd.pivot_table(data,index=['What is your gender?'],values=basic_params,aggfunc='mean').T
```

```python
graph = basic_param_avg.plot(kind='bar',figsize=(16,6),title='Avg. Importance of Basic parameters')
for p in graph.patches:
    graph.annotate('{:.2f}'.format(p.get_height()), (p.get_x()+0.12, p.get_height()),
                   ha='center', va='bottom', color= 'black')
```

```python
genders = data['What is your gender?'].unique()
```

```python
basic_param_avg.reset_index(inplace=True)
basic_param_avg.rename(columns={'index':'parameters'},inplace=True)
```

```python
for gender in genders:
    basic_param_avg.sort_values(by=gender,inplace=True)
    plt.figure(figsize=(16,7))
    gender_basic = sns.barplot(data=basic_param_avg,x='parameters',y=gender,palette='coolwarm')
    gender_basic.set(ylabel='Importance (out of 5)')
    gender_basic.set_title('Basic Parameter average importance - '+gender,fontsize=15)
    gender_basic.set_xticklabels(gender_basic.get_xticklabels(), 
                                     rotation=45, horizontalalignment='right')

    for p in gender_basic.patches:
        gender_basic.annotate('{:.2f}'.format(p.get_height()), (p.get_x()+0.4, p.get_height()),
                                  ha='center', va='bottom', color= 'black')
```

```python
lifestyle_param_avg = pd.pivot_table(data,index=['What is your gender?'],values=lifestyle_params,aggfunc='mean').T
```

```python
graph = lifestyle_param_avg.plot(kind='bar',figsize=(16,7),cmap='Accent',
                                 title='Avg. Importance of Lifestyle parameters')
for p in graph.patches:
    graph.annotate('{:.2f}'.format(p.get_height()), (p.get_x()+0.12, p.get_height()),
                   ha='center', va='bottom', color= 'black')
```

```python
lifestyle_param_avg.reset_index(inplace=True)
lifestyle_param_avg.rename(columns={'index':'parameters'},inplace=True)
```

## Lifestyle Parameter average importance(Gender wise)

```python
for gender in genders:
    lifestyle_param_avg.sort_values(by=gender,inplace=True)
    plt.figure(figsize=(16,7))
    gender_lifestyle = sns.barplot(data=lifestyle_param_avg,x='parameters',y=gender,palette='icefire')
    gender_lifestyle.set(ylabel='Importance (out of 5)')
    gender_lifestyle.set_title('Lifestyle Parameter average importance - '+gender,fontsize=15)
    gender_lifestyle.set_xticklabels(gender_lifestyle.get_xticklabels(), 
                                     rotation=45, horizontalalignment='right')

    for p in gender_lifestyle.patches:
        gender_lifestyle.annotate('{:.2f}'.format(p.get_height()), (p.get_x()+0.4, p.get_height()),
                                  ha='center', va='bottom', color= 'black')
```
